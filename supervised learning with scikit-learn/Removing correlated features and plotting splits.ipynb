```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(
    r'D:\Code-files\Visual code\python for machine learning\datasets\BCancer.csv')
df = df.drop(columns=['Unnamed: 32'], axis=1)
df
```

```python
def remove_corr(df, threshold):  
    cor_matrix = df.corr()
    upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(np.bool))
    to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > threshold)]
    df1 = df.drop(to_drop, axis=1)
    return df1

df.corr()
```

```python
df = remove_corr(df, 0.8)
df.corr() #Correlation after removal of features
```

```python
df
```

```python
X = df.iloc[:, 2:]
Y = df['diagnosis']

from sklearn.model_selection import train_test_split as t_split
x1_train, x1_test, y1_train, y1_test = t_split(X, Y, test_size=.3)
x2_train, x2_test, y2_train, y2_test = t_split(X, Y, test_size=.5)
x3_train, x3_test, y3_train, y3_test = t_split(X, Y, test_size=.6)


from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Finding best value for K, by iterating through and finding best accuracy for corresponding K value
def getAccuracy(X, Y, size):
    scores = []
    k_vals = []
    highest_accuracy = 0 
    best_k = 0
    
    for i in range(3, 21, 2):
        classifier = KNeighborsClassifier(n_neighbors=i)
        x1_train, x1_test, y1_train, y1_test = t_split(X, Y, test_size=size)
        classifier.fit(x1_train, y1_train)
        y_pred = classifier.predict(x1_test)
        accuracy = accuracy_score(y1_test, y_pred)
        scores.append(accuracy)
        if round(highest_accuracy, 2) < round(accuracy, 2):
            heighest_accuracy = accuracy
            best_k = i
        k_vals.append(i)
        
    return k_vals, scores, best_k
```

```python
# Splitting each dataset, with specified train-test size
k_vals, scores1, best_k1 = getAccuracy(X, Y, 0.3)
k_vals, scores2, best_k2 = getAccuracy(X, Y, 0.5)
k_vals, scores3, best_k3 = getAccuracy(X, Y, 0.6)

# print(f'Best value for 1st split: {best_k1},\nBest value for 2nd split: {best_k2},\nBest value for 3rd split: {best_k3}')
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(10, 5))
plt.subplot
plt.subplots(1, 1, figsize=(10, 5))
plt.plot(k_vals, scores1, label='Split 70-30', linewidth=3, color='green')
plt.title('Split 70-30')
plt.xlabel('K Value')
plt.ylabel('Accuracy')

plt.subplots(1, 1, figsize=(10, 5))
plt.plot(k_vals, scores2, label='Split 50-50', linewidth=3, color='cyan')
plt.title('Split 50-50')
plt.xlabel('K Value')
plt.ylabel('Accuracy')

plt.subplots(1, 1, figsize=(10, 5))
plt.plot(k_vals, scores3, label='Split 40-60', linewidth=3, color='red')
plt.title('Split 40-60')
plt.xlabel('K Value')
plt.ylabel('Accuracy')

plt.tight_layout()
plt.show()
```

```python
# Finding best value using variance approach
def variance_calc(X, Y, size):
    train_score = []
    test_score = []
    k_vals = []
    variance = []

    for k in range(3, 21, 2):
        x_train, x_test, y_train, y_test = t_split(X, Y, test_size=size)
        knn = KNeighborsClassifier(n_neighbors=k)
        k_vals.append(k)
        knn.fit(x_train, y_train)

        train_k_score = knn.score(x_train, y_train)
        train_score.append(train_k_score)
        test_k_score = knn.score(x_test, y_test)
        test_score.append(test_k_score)

        variance_val = abs(train_k_score - test_k_score)
        variance.append(variance_val)
    return variance, k_vals
```

```python
variance1, k_vals = variance_calc(X, Y, .3)
variance2, k_vals = variance_calc(X, Y, .5)
variance3, k_vals = variance_calc(X, Y, .6)

plt.figure(figsize=(20, 10))
plt.style.use('fivethirtyeight')

plt.subplot
plt.subplots(1, 1, figsize=(10, 5))
plt.plot(k_vals, variance1, linewidth=3)
plt.subplots(1, 1, figsize=(10, 5))
plt.plot(k_vals, variance2, linewidth=3)
plt.subplots(1, 1, figsize=(10, 5))
plt.plot(k_vals, variance3, linewidth=3)

plt.xlabel('K Value')
plt.ylabel('Variance')
plt.tight_layout()
plt.show()
```
